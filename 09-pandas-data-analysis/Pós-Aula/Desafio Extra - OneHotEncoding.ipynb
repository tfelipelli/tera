# Série Original:


````python
classes = ["Leite", "Ovos", "Carne", "Arroz", "Feijão"]
labels = pd.Series(np.random.choice(classes, 100))
````

# Dataset Codificado:

##### Solução 1:

````python
df = pd.DataFrame(
    index=labels.index,
    columns=labels.unique(),
    data=0.
)
for k in classes:
    df.loc[labels == k, k] = 1.
df
````

##### Solução 2:

````python
df = pd.DataFrame(index=labels.index)

for k in classes:
    df[k] = (labels == k).astype(float)
df
````
